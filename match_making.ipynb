```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from faker import Faker
import random

fake = Faker()


num_students_large = 1000

data_large = {
    'student_id': [str(i) for i in range(1, num_students_large + 1)],
    'student_name': [fake.name() for _ in range(num_students_large)],
    'interest': [random.choice(['Math', 'Art', 'Science', 'History']) for _ in range(num_students_large)],
}


df_large = pd.DataFrame(data_large)


df_large.to_csv('large_student_interests.csv', index=False)


df_large = pd.read_csv('large_student_interests.csv')


interest_matrix_large = pd.get_dummies(df_large['interest'])


df_large = pd.concat([df_large, interest_matrix_large], axis=1)


df_large = df_large.drop('interest', axis=1)


X_large = df_large.iloc[:, 2:].values  
similarity_matrix_large = cosine_similarity(X_large)


def find_similar_students_large(student_id, top_n=3):
    try:
        student_index = df_large[df_large['student_id'] == student_id].index.values[0]
        similarities = similarity_matrix_large[student_index]

        
        similar_students_indices = similarities.argsort()[:-top_n-1:-1][1:]
        
        
        similar_students_ids = df_large.loc[similar_students_indices, 'student_id'].tolist()
        
        return similar_students_ids
    except IndexError:
        print(f"Error: Student ID {student_id} not found in the large dataset.")
        return []


random_student_id = random.choice(df_large['student_id'].tolist())
similar_students_large = find_similar_students_large(random_student_id, top_n=4)
print(f"Similar Students for Random Student ID {random_student_id} in Large Dataset:", similar_students_large)
```
